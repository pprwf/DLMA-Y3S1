```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount("/content/drive")
```

```python
img = cv2.imread("/content/drive/My Drive/lena.jpg", 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap = "gray")
```

**Lab 01: Horizontal Flip**

```python
img = img[::, ::-1]

plt.axis("off")
plt.imshow(img, cmap = "gray")
```

**Lab 02: Make a black cencorship on Lena's eyes**

```python
img[105:130, 65:130] = 0

plt.axis("off")
plt.imshow(img, cmap = "gray")
```

**Lab 03: Resize picture from 256x256 into 128x128**

```python
img = img[::2, ::2]


plt.imshow(img, cmap = "gray")
```
